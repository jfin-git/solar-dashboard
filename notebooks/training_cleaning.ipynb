```python
import pandas as pd

df = pd.read_csv(r'C:\Users\jsfin\OneDrive\Desktop\Projects\solar-dashboard\data\processed\Farmington_final_training.csv', parse_dates=['timestamp'], index_col='timestamp')
```

```python
# Visualize temperature data over time
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Temperature'], label='Temperature (°C)', color='orange')
plt.plot(df.index, df['Relative Humidity'], label='Relative Humidity', color='blue')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
# Visualize Clearsky GHI over time
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Clearsky GHI'], label='Clearsky GHI (W/m²)', color='blue')
plt.xlabel('Time')
plt.ylabel('Clearsky GHI (W/m²)')
plt.title('Clearsky GHI Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
# Visualize Solar Zenith Angle over time
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Solar Zenith Angle'], label='Solar Zenith Angle (degrees)', color='green')
plt.xlabel('Time')
plt.ylabel('Solar Zenith Angle (degrees)')
plt.title('Solar Zenith Angle Over Time')
plt.legend()
plt.grid()
plt.show()
```

```python
# Visualize target_mw over time
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['target_mw'], label='Target MW', color='red')
plt.xlabel('Time')
plt.ylabel('Target MW')
plt.title('Target MW Over Time')
plt.legend()
plt.grid()
plt.show()
```
